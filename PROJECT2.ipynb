```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
location = "datasets/DC_Properties.csv"
df = pd.read_csv(location)
```

```python
#make a scatterplot
sns.lmplot(x='ROOMS', y='PRICE', data=df)
```

```python
#add more details to scatterplot
sns.lmplot(x='ROOMS', y='PRICE', data=df, 
           fit_reg=False, #remove regression line
           hue='QUADRANT') 
```
